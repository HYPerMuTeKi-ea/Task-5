```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Create figure with unit square and inscribed quarter-circle
plt.figure()
# Draw quarter circle with unit radius
theta_circ = np.linspace(0,np.pi/2,1001)
x_circ     = np.cos(theta_circ)
y_circ     = np.sin(theta_circ)
plt.plot(x_circ,y_circ,'-b')
plt.gca().set_aspect('equal') # set aspect ratio so shape not distorted


plt.plot([0,1,1,0,0],[0,0,1,1,0],'b')

# Generate random arrays of x and y coordinates within unit square
n_points = 1000
x_points = np.random.rand(n_points)
y_points = np.random.rand(n_points)

# Plot generated points
plt.plot(x_points,y_points,'.')
```

```python
n_inside = 0
i=0
while i<n_points:
    x = np.random.rand()
    y = np.random.rand()
    if (x**2 + y**2) <= 1:
        n_inside += 1
    i+=1
pi_approx = 4 * n_inside / n_points
print(f"Of {n_points} points, {n_inside} are inside the quarter circle.")
print(f"Estimate of pi = {pi_approx}.")
```

```python
n_values = 1000
i=0
pi_values = np.empty(n_values+1)
while i<=n_values:
    j=0
    n_inside = 0
    while j< n_points:
        x = np.random.rand()
        y = np.random.rand()
        if (x**2 + y**2) <= 1:
            n_inside += 1
        j+=1
    
    pi_approx = 4 * n_inside / n_points
    pi_values[i] = pi_approx
    print(f"Estimate of pi = {pi_approx:.3f}")
    i+=1
plt.rcParams['patch.force_edgecolor'] = True
plt.figure()
plt.hist(pi_values);
```
